```python
!pip install pandas scikit-learn

# Import required libraries
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load datasets
customers = pd.read_csv(customers_url)
products = pd.read_csv(products_url)
transactions = pd.read_csv(transactions_url)

# Merge datasets
merged = transactions.merge(customers, on='CustomerID', how='left')
merged = merged.merge(products, on='ProductID', how='left')

# Aggregate customer-product interaction
customer_product_features = merged.groupby(['CustomerID', 'ProductID']).agg(
    {'Quantity': 'sum', 'TotalValue': 'sum'}).reset_index()

# Pivot table to create a user-item matrix
customer_product_matrix = customer_product_features.pivot(index='CustomerID',
                                                           columns='ProductID',
                                                           values='TotalValue').fillna(0)

# Normalize matrix
normalized_matrix = customer_product_matrix.div(customer_product_matrix.sum(axis=1), axis=0)

# Calculate similarity scores
similarity_scores = cosine_similarity(normalized_matrix)

# Create a DataFrame for similarity scores
similarity_df = pd.DataFrame(similarity_scores, index=customer_product_matrix.index,
                             columns=customer_product_matrix.index)

# Find top 3 lookalikes for the first 20 customers
lookalike_results = {}

for customer in similarity_df.index[:20]:
    top_matches = similarity_df[customer].nlargest(4).iloc[1:]
    lookalike_results[customer] = [(match, score) for match, score in top_matches.items()]

# Create Lookalike.csv
lookalike_df = pd.DataFrame.from_dict(lookalike_results, orient='index',
                                      columns=['Lookalike1', 'Lookalike2', 'Lookalike3'])
lookalike_df.to_csv('Lookalike.csv', index_label='CustomerID')

# Output lookalike results
print(lookalike_df.head())
```

```python
# Task 3: Customer Segmentation
# Install necessary libraries
!pip install pandas scikit-learn matplotlib seaborn

# Import required libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
customers_url = "https://drive.google.com/uc?id=1bu_--mo79VdUG9oin4ybfFGRUSXAe-WE"
products_url = "https://drive.google.com/uc?id=1IKuDizVapw-hyktwfpoAoaGtHtTNHfd0"
transactions_url = "https://drive.google.com/uc?id=1saEqdbBB-vuk2hxoAf4TzDEsykdKlzbF"

customers = pd.read_csv(customers_url)
products = pd.read_csv(products_url)
transactions = pd.read_csv(transactions_url)

# Merge datasets
merged = transactions.merge(customers, on='CustomerID', how='left')
merged = merged.merge(products, on='ProductID', how='left')

# Verify merged data
print(merged.head())
print("Columns in merged dataset:", merged.columns)

# Add a default price if missing after merging
if 'Price' not in merged.columns or merged['Price'].isnull().all():
    print("Price column is missing or entirely null. Setting default price to 0.")
    merged['Price'] = 0.0  # Assign a default price if missing

# Feature selection for clustering
clustering_data = merged.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum',
    'Price': 'mean'  # This is now guaranteed to exist
}).reset_index()

# Normalize data
scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data.drop(columns=['CustomerID']))

# K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(clustering_data_scaled)
clustering_data['Cluster'] = kmeans.labels_

# Evaluate clusters
db_index = davies_bouldin_score(clustering_data_scaled, kmeans.labels_)
print(f"Davies-Bouldin Index: {db_index}")

# Visualize clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=clustering_data_scaled[:, 0], y=clustering_data_scaled[:, 1],
                hue=clustering_data['Cluster'], palette='viridis', s=100)
plt.title("Customer Segmentation Clusters")
plt.xlabel("Feature 1 (Scaled)")
plt.ylabel("Feature 2 (Scaled)")
plt.show()

# Save clustering results
clustering_data.to_csv('Customer_Segmentation.csv', index=False)
print("Clustering results saved to Customer_Segmentation.csv")
```
